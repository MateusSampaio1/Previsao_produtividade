Antes de decidir qual será a abordagem escolhida para tratar os dados faltantes da coluna WIP, acredito que seja interessante descrever o conteúdo de cara coluna.


- **date** - Data no formato Mês - Dia - Ano;
- **quarter** - Uma fração parte do mês, aqui o mês foi dividido em 3;
- **department** - O departamento;
- **day** - Dia da semana que o dado foi coletado;
- **team** - Número associado ao time;
- **targeted_productivity** - Produtividade desejada;
- **smv** - tempo alocado para cada tarefa (padrão);
- **wip** - Trabalho em desenvolvimento, inclui a qtd de itens inacabados p/ produto;
- **over_time** - horas extras de cada equipe;
- **incentive** - Valor do incentivo financeiro (em BDT);
- **idle_time** - Tempo ocioso (produção parada por qualquer motivo);
- **idle_men** - Qtd de trabalhadores que ficarão ociosos devido a parada na produção;
- **no_of_style_change** - Número de mudanças no estilo de um determinado produto;
- **no_of_workers** - Número de trabalhadores por equipe
- **actual_productivity** - A porcentagem da produção que foi realmente entregue

> Como pode-se observar existe uma grande quantidade de registros nulos na coluna **wip** que se refere ao trabalho em desenvolvimento, incluindo os produtos inacabados, esses registros correspondem a quase metade ddo Dataset e por isso aredito que não será uma boa ideia remover essas linhas.

```python
df['wip'].describe()
```

> Existe uma pequena parcela dos dados que contêm uma grande qtd de itens inacabados, distorcendo a média e tornando inviável utilizá-la para substituir esses dados fantantes. Também poderia utilizar a mediana, no entanto, como essa coluna aborda os itens que ainda não estão terminados, irei considerar que esses dados não foram preenchidos pois no momento da coleta dos dados não existiam itens inacabados. 

> Logo irei subistituir essa linhas com 0.

```python
# Colocando 0 nas linhas com dados faltantes da coluna wip
df['wip'].fillna(0, inplace  = True)
```

```python
# Arrumando o erro entrado nos dados da coluna departament
df['department'].replace({'finishing ':'finishing'}, inplace = True)
```

Será necessário realizar o enconding dessas variáveis categóricas para que seja possível utilizá-las no modelo preditivo. Esse etapa será realizada atravez da ferramenta get_dummies do pandas.

```python
# Transformando as variáveis categórias em numéricas
df = pd.get_dummies(data = df, columns = ['quarter','department','day'])
```

```python
# Reorganizando as colunas 
df = df[['date','day_Sunday','day_Monday','day_Tuesday','day_Wednesday','day_Thursday','day_Saturday','quarter_Quarter1','quarter_Quarter2','quarter_Quarter3','quarter_Quarter4','quarter_Quarter5',
    'department_finishing','department_sweing','team','targeted_productivity','smv','wip','over_time','incentive','idle_time','idle_men','no_of_style_change','no_of_workers','actual_productivity']]
df.head()
```

```python
X = df.drop(['actual_productivity','date'],axis = 1)
y = df['actual_productivity']
```

```python
from sklearn.preprocessing import StandardScaler

scaler =  StandardScaler()
X_new = scaler.fit_transform(X)
X_new = pd.DataFrame(X_new, columns = X.columns)
X_new
```

```python
pip install xgboost
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# Definindo uma SEED
SEED = 15973
np.random.seed(SEED)

# Separando os dados entre treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size= 0.3, 
                                                   random_state = SEED, shuffle= True)
```

```python
# Treinando o modelo RandomForestRegressor
modelo_rfr = RandomForestRegressor(random_state = SEED)

#Ajustando o modelo
modelo_rfr.fit(X_train, y_train)

# Realizando a previsão 
y_pred_rfr = modelo_rfr.predict(X_test)

# Métricas para o modelo
r2 = r2_score(y_test,y_pred_rfr) *100
r2_ajustado = (1-((1- r2_score(y_test,y_pred_rfr))*(len(y_test)-1))/(len(y_test) - X_train.shape[1] -1)) * 100
RMSE = (np.sqrt(mean_squared_error(y_test, y_pred_rfr)))
MAE = mean_absolute_error(y_test, y_pred_rfr)

print('O R² deste modelo é de %.4f%%'%r2)
print('O R² ajustado deste modelo foi de %.4f%%'%r2_ajustado)
print('A raiz do erro quadrático médio deste modelo foi de %.4f'%RMSE)
print('O erro absoluto médio deste modelo foi de %.4f'%MAE)
```

```python
# Treinando o modelo XGBRegressor
modelo_xgb = XGBRegressor(random_state = SEED)

#Ajustando o modelo
modelo_xgb.fit(X_train, y_train)

# Realizando a previsão 
y_pred_xgb = modelo_xgb.predict(X_test)

# Métricas para o modelo
r2 = r2_score(y_test,y_pred_xgb) *100
r2_ajustado = (1-((1- r2_score(y_test,y_pred_xgb))*(len(y_test)-1))/(len(y_test) - X_train.shape[1] -1)) * 100
RMSE = (np.sqrt(mean_squared_error(y_test, y_pred_xgb)))
MAE = mean_absolute_error(y_test, y_pred_xgb)

print('O R² deste modelo é de %.4f%%'%r2)
print('O R² ajustado deste modelo foi de %.4f%%'%r2_ajustado)
print('A raiz do erro quadrático médio deste modelo foi de %.4f'%RMSE)
print('O erro absoluto médio deste modelo foi de %.4f'%MAE)
```

>Como as métricas relativas ao erro absoluto (MAE) e a raiz do erro quadrático médio (RMSE) foram bem próximas, irei tentar otimizar o modelo XGBRegressor.

```python
# Instanciando novamento o modelo XGBRegressor

modelo = XGBRegressor(n_estimators = 500,
                      learning_rate = 0.01,
                      max_depth =5,
                      min_child_weight = 1,
                      subsample = 0.85,
                      colsample_bytree=0.96,
                      random_state = SEED)

modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)

RMSE = (np.sqrt(mean_squared_error(y_test, y_pred)))
MAE = mean_absolute_error(y_test, y_pred)

print('A raiz do erro quadrático médio deste modelo foi de %.4f'%RMSE)
print('O erro absoluto médio deste modelo foi de %.4f'%MAE)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

# Instanciando o objeto KFold e atribuindo a variável cv
cv = KFold(n_splits = 10, shuffle= True, random_state = SEED)

# Realizando a validação cruzada e  obtendo o score do erro quadrático médio
score = cross_val_score(modelo, X_new, y, scoring = 'neg_mean_squared_error', cv = cv)

# Como a ideia é obter o intervalo para o RMSE irei tirar a raiz do score antes de calcular a média e o desvio padrão

RMSE_score = np.sqrt(-score)

#Obtendo a média e o desvio padão

media = RMSE_score.mean()
desv_pad = RMSE_score.std()

print('A média do RMSE para esse modelo é de %.4f'%media)
print('O intervalo esperado é de [%.4f, %.4f]'%(media - 2*desv_pad, media + 2*desv_pad))
```

Nesta ocasião em específico a otimização do modelo foi feita de maneira manual, testando parâmetro a parâmetro individualmente ou em conjunto, e ainda assim foi possível encontrar um resultado melhor que o inicial do RandomForest.

Como é possível observar a média é bem proxima aos 0.12 encontrado depois da otimização e o intervalo também 
Acredito que seja possível obter um RMSE menor utilizando uma RandomizedSearch ou uma bayesian optmazation.
